And defined some helper functions to vizualise our data.

```python
# Time Series on classes:
def load_sound_files(file_paths):
    raw_sounds = []
    for fp in file_paths:
        X,sr = librosa.load(fp)
        raw_sounds.append(X)
    return raw_sounds

def plot_waves(sound_names,raw_sounds):
    i = 1
    fig = plt.figure(figsize=(35,15))

    for n,f in zip(sound_names,raw_sounds):
        plt.subplot(2,5,i)
        librosa.display.waveshow(np.array(f),sr=22050, color='orange', alpha=0.7,where = 'post')
        plt.title(n.title())
        i += 1

    fig.suptitle('Figure 1: Waveplot, Time Series', fontsize=18)
    #plt.savefig("time_series")
    plt.tight_layout()
    plt.show()

def plot_spec(sound_names,raw_sounds):
    i = 1
    fig = plt.figure(figsize=(35,15))

    for n,f in zip(sound_names,raw_sounds):
        plt.subplot(2,5,i)
        d =librosa.amplitude_to_db(np.abs(librosa.stft(f)), ref=np.max)
        librosa.display.specshow(d, y_axis='linear')
        plt.colorbar(format='%+2.0f dB')
        plt.title(n.title())
        i += 1

    fig.suptitle('Figure 2: Linear-Frequency Power Spectrogram', fontsize=18)
    #plt.savefig("spectrogram")
    plt.tight_layout()
    plt.show()
```

## Extraction

We analysed different features we could extract, such as:
- STFTs
- MFCCs
- Melspectograms
- Chromagrams
- Constant-Q Chromagrams
- CENS
- Filter Bank Coefficients

Each shape on a dog bark labelled sample:

```python
# Feature set:
# This file is of a dog bark:

y,sr = librosa.load(df['file_path'][0])

# stft -> Short-time Fourier transform:(The STFT represents a signal in the time-frequency domain by computing discrete Fourier transforms (DFT) over short overlapping windows)
S = librosa.stft(y)
# Mel-frequency cepstral coefficients:
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
# Compute a mel-scaled spectrogram:
melspectrogram =librosa.feature.melspectrogram(y=y, sr=sr, n_mels=40,fmax=8000)
# Compute a chromagram from a waveform or power spectrogram:
chroma_stft=librosa.feature.chroma_stft(y=y, sr=sr,n_chroma=40)
# Constant-Q chromagram:
chroma_cq =librosa.feature.chroma_cqt(y=y, sr=sr, n_chroma=12)
# Computes the chroma variant “Chroma Energy Normalized” (CENS):
chroma_cens =librosa.feature.chroma_cens(y=y, sr=sr, n_chroma=12)
# Filter Bank Coefficients:
filter_bank = librosa.filters.semitone_filterbank()

print(f"Melspectrogram Shape : {melspectrogram.shape}")
print(f"MFCC Shape : {mfccs.shape}")
print(f"Chromagram Shape : {chroma_stft.shape}")
print(f"Constant Q Chromagram Shape : {chroma_cq.shape}")
print(f"CENS Shape : {chroma_cens.shape}")
```

```python
def wave_plot(filename,class_label):
    """
    Plot the waveform of a given WAV audio file and display its metadata.

    Parameters:
    - filename (str): The path to the WAV audio file.
    - class_label (str): A label or category for the audio file.

    Returns:
    None
    """
    # Read WAV file using scipy.io.wavfile
    rate, wav_audio = wav.read(filename)

    # Open WAV file for additional metadata extraction          
    wave_file = open(filename,"rb")

    # Read the first 36 bytes (which include metadata) from the WAV file
    riff_fmt = wave_file.read(36)

    # Extract bit depth from the last 2 bytes of the metadata
    bit_depth_string = riff_fmt[-2:]
    bit_depth = struct.unpack("H",bit_depth_string)[0]

    # Display audio metadata
    print(f"Sampling Rate: {rate} Hz\nBit Depth: {bit_depth}\nNumber of Channels: {wav_audio.shape[1]}\nDuration: {wav_audio.shape[0]/rate} second\nNumber of Samples: {len(wav_audio)}\nClass: {class_label}")

    # Plot the waveform using matplotlib
    plt.figure(figsize=(12, 4))
    plt.plot(wav_audio)
```

This is an example usage of the `wave_plot()` function on a sample.

Finally, we decided to extract the MFCCs and the Chromagram. After that we padded each feature with 0s in order to have an uniform sample size.

```python
max_pad_len = 174

def extract_features(file_name):

    try:
        audio, sample_rate = librosa.load(file_name)
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
        chroma_stft=librosa.feature.chroma_stft(y=audio, sr=sample_rate,n_chroma=40)
        # feature scaling:
        pad_width = max_pad_len - mfccs.shape[1]
        mfccs = np.pad(mfccs, pad_width=((0, 0), (0, pad_width)), mode='constant')
        pad_width = max_pad_len - chroma_stft.shape[1]
        chroma_stft = np.pad(chroma_stft, pad_width=((0, 0), (0, pad_width)), mode='constant')

    except Exception as e:
        print("Error encountered while parsing file: ", file_name)
        return None

    return mfccs, chroma_stft
```

```python
X_mfcc = [ [] for _ in range(10) ]
X_chroma = [ [] for _ in range(10) ]
Y = [ [] for _ in range(10) ]

# Iterate through each sound file and extract the features:
for index, row in df.iterrows():
    if (index+1)%100 == 0:
        print(f"File {index+1} out of {len(df)}")


    file_name = row["file_path"]

    class_id = row["classID"]
    mfcc, chroma  = extract_features(file_name)
    X_mfcc[row["fold"]-1].append(mfcc)
    X_chroma[row["fold"]-1].append(chroma)
    Y[row["fold"]-1].append(class_id)

#   write pickled data to file
with open('X_mfcc.pkl', 'wb') as f:
    pickle.dump(X_mfcc, f)

with open('X_chroma.pkl', 'wb') as f:
    pickle.dump(X_chroma, f)

with open('Y.pkl', 'wb') as f:
    pickle.dump(Y, f)
```
